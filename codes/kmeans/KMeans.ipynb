```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SEED = 73

np.random.seed(SEED)
```

```python
def get_kmeans_data():

    # 1. Get Data
    data_filepath = "../../data/kmeans/kmeans_data/data.npy"
    labels_filepath = "../../data/kmeans/kmeans_data/label.npy"

    data = np.load(data_filepath)
    labels = np.load(labels_filepath)

    assert len(data) == len(labels)

    # 2. Normalize Data
    means = data.mean(axis=0)
    stds = data.std(axis=0)

    normalization_values = {"mean": means, "std": stds}

    data = (data - means) / stds

    # 3. Shuffle Data
    indices = list(range(len(data)))
    np.random.shuffle(indices)

    data = data[indices]
    labels = labels[indices]

    print("Data Shape    : {}".format(data.shape))
    print("Labels Shape  : {}".format(labels.shape))
    print("Unique Labels : {}".format(np.unique(labels)))

    print("\nNormalization Values : {}".format(normalization_values))

    print("\nData Mean : {}".format(data.mean(axis=0)))
    print("Data Std  : {}".format(data.std(axis=0)))
    
    return data, labels, normalization_values
```

```python
def plot_data(data, labels):
    
    plt.figure(figsize=(10, 10))
    plt.title("Data Points")
    plt.scatter(data[:, 0], data[:, 1] , c=labels)
    plt.grid()
    plt.show()
```

```python
data_normalized, labels, normalization_values = get_kmeans_data()
n_classes = len(np.unique(labels))
```

```python
plot_data(data_normalized, labels)
```

```python
class KMeansModel(object):

    def __init__(self, n_clusters, n_iterations):

        self.n_clusters = n_clusters
        self.n_iterations = n_iterations

        self.centroids = None

    def __calculate_euclidean_distances(self, sample, targets):
        
        return np.sqrt(np.sum((sample - targets) ** 2, axis=1))

    def __initialize_centroids(self, data):

        centroids = [data[np.random.choice(list(range(len(data))))], ]

        for _ in range(self.n_clusters - 1):
            centroid_distances = np.array([self.__calculate_euclidean_distances(cnt, data) for cnt in centroids])
            centroid_distances_sum = np.sum(centroid_distances, axis=0)
            
            centroid_distances_sum_norm = centroid_distances_sum / np.sum(centroid_distances_sum) # as prob. density

            sel_centroid_idx = np.random.choice(list(range(len(data))), p=centroid_distances_sum_norm)
            sel_centroid = data[sel_centroid_idx]

            centroids.append(sel_centroid)

        centroids = np.array(centroids)

        return centroids
    
    def fit(self, data):

        self.centroids = self.__initialize_centroids(data)

        assert len(self.centroids) == self.n_clusters

        for iter_idx in range(self.n_iterations):

            prev_centroids = self.centroids.copy()

            assignments = [[] for _ in range(self.n_clusters)]

            for datum_idx, datum in enumerate(data):
                distances = self.__calculate_euclidean_distances(datum, self.centroids)
                min_dist = np.min(distances)

                sel_centroid_indices = np.where([dist == min_dist for dist in distances])[0]
                sel_centroid_idx = np.random.choice(sel_centroid_indices)

                assignments[sel_centroid_idx].append(datum_idx)

            new_centroids = self.centroids.copy()

            for centroid_idx, assignment in enumerate(assignments):
                if len(assignment) != 0:
                    new_centroids[centroid_idx] = np.mean([data[datum_idx] for datum_idx in assignment], axis=0)

            if np.all(prev_centroids == new_centroids):
                break

            self.centroids = new_centroids

        print("Final Iteration : {}".format(iter_idx))

    def predict(self, data):

        predictions = []
        
        for datum in data:
            distances = self.__calculate_euclidean_distances(datum, self.centroids)
            sel_centroid_idx = np.argmin(distances)

            predictions.append(sel_centroid_idx)

        return predictions
```

```python
kmeans_model = KMeansModel(n_clusters=n_classes, n_iterations=100)
kmeans_model.fit(data_normalized)

print("Centroids :\n{}".format(kmeans_model.centroids))
```

```python
predictions = kmeans_model.predict(data_normalized)

plot_data(data_normalized, predictions)
```
